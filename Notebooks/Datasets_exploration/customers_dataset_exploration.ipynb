```python
states_count= customers["customer_state"].value_counts()
```

```python
plt.figure(figsize = (12,6))
sns.barplot(x = states_count.index, y = states_count.values)

plt.title("Number of order per State")
plt.xlabel("State")
plt.ylabel("Nomber of Order")
plt.xticks(rotation = 45)

plt.show
```

```python
city_count = customers["customer_city"].value_counts().sort_values(ascending = False).head(30)
```

```python
plt.figure(figsize= (25,6))
sns.barplot(x = city_count.index, y = city_count.values)

plt.title("Number of order per city")
plt.xlabel("City")
plt.ylabel("Number of orders")
plt.xticks(rotation = 270)

plt.show()
```
