```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customers = pd.read_csv("../../Datasets_Olist/olist_customers_dataset.csv")
customers.head(5)
```

```python
states_count= customers["customer_state"].value_counts()
```

```python
plt.figure(figsize = (12,6))
sns.barplot(x = states_count.index, y = states_count.values)

plt.title("Number of order per State")
plt.xlabel("State")
plt.ylabel("Nomber of Order")
plt.xticks(rotation = 45)

plt.show
```

```python
city_count = customers["customer_city"].value_counts().sort_values(ascending = False).head(30)
```

```python
plt.figure(figsize= (25,6))
sns.barplot(x = city_count.index, y = city_count.values)

plt.title("Number of order per city")
plt.xlabel("City")
plt.ylabel("Number of orders")
plt.xticks(rotation = 270)

plt.show()
```

```python
customers_per_state = customers.groupby("customer_state")["customer_unique_id"].nunique().sort_values(ascending= False)
```

```python
plt.figure(figsize= (16,8))
sns.barplot(x= customers_per_state.index, y=customers_per_state.values)

plt.title("Number of customers per state")
plt.xlabel("State")
plt.ylabel("Number of customers")

plt.show()
```

```python
customers_per_city = customers.groupby("customer_city")["customer_unique_id"].nunique().sort_values(ascending= False).head(30)
```

```python
plt.figure(figsize= (16,8))
sns.barplot(x= customers_per_city.index, y=customers_per_city.values)

plt.title("Number of customers per City")
plt.xlabel("City")
plt.ylabel("Number of customers")
plt.xticks(rotation = 270)

plt.show()
```

```python
number_of_order_customers = customers.groupby("customer_unique_id")["customer_id"].nunique().sort_values(ascending= False).head(30)
```

```python
plt.figure(figsize= (16,8))
sns.barplot(x= number_of_order_customers.index, y=number_of_order_customers.values)

plt.title("Number of orders per customers")
plt.xlabel("customers")
plt.ylabel("Number of order")
plt.xticks(rotation = 270)

plt.show()
```
