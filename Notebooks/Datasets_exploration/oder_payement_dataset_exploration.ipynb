```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
order_payement = pd.read_csv('../../Datasets_Olist/olist_order_payments_dataset.csv')
order_payement.head(5)
```

```python
plt.scatter(range(len(order_payement)), order_payement['payment_value'])
plt.title("Scatter plot for order payment outlier")
plt.show()
```

```python
order_per_payment_type = order_payement["payment_type"].value_counts()

plt.figure(figsize= (16,8))
sns.barplot(x= order_per_payment_type.index, y=order_per_payment_type.values)

plt.title("Number of order per payment type")
plt.xlabel("Payment Type")
plt.ylabel("Number of order")
plt.xticks(rotation = 270)

plt.show()
```

```python
order_per_payment_installement = order_payement.groupby("payment_installments")["order_id"].nunique()

plt.figure(figsize= (16,8))
sns.barplot(x= order_per_payment_installement.index, y=order_per_payment_installement.values)

plt.title("Number of order per payment installment")
plt.xlabel("Payment installment")
plt.ylabel("Number of order")
plt.xticks(rotation = 270)

plt.show()
```
