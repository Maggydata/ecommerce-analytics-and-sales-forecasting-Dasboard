```python
product_most_like = order_items.groupby("product_id")["order_id"].nunique().sort_values(ascending= False).head(30)
```

```python
plt.figure(figsize= (16,8))
sns.barplot(x= product_most_like.index, y=product_most_like.values)

plt.title("Products the most sold")
plt.xlabel("Product")
plt.ylabel("Number of Order")
plt.xticks(rotation = 270)

plt.show()
```
