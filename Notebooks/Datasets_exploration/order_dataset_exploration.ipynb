```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
order = pd.read_csv("../../Datasets_Olist/olist_orders_dataset.csv")
order.head(5)
```

```python
order.shape
```

```python
order["order_status"].unique()
```

```python
order_per_status = order["order_status"].value_counts()

plt.figure(figsize= (16,8))
sns.barplot(x= order_per_status.index, y=order_per_status.values)

plt.title("Number of order per status")
plt.xlabel("Status")
plt.ylabel("Number of order")
plt.xticks(rotation = 270)

plt.show()
```
